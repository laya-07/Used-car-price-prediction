# Error Analysis

The model performed well on typical used cars and generated realistic predictions. However, some limitations were observed:

1. The model may be less accurate for luxury or premium vehicles because such vehicles are underrepresented in the training dataset.

2. Predictions may occasionally be slightly higher than the present price of a vehicle. This occurs because the ElasticNet regression model learns statistical patterns from the data and does not enforce business constraints.

3. The model's performance depends on the quality and diversity of the training data. Vehicles with unusual characteristics may result in larger prediction errors.

4. Market factors such as brand reputation, vehicle condition, accident history, and regional demand are not included in the dataset and may affect real-world selling prices.

Overall, the model achieved a Test R² score of 0.8518, indicating strong predictive performance despite these limitations.
